```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def find_first_nan_row(df):
    nan_ix = df.index[df.isna().all(axis=1)][0]
    return nan_ix
```

```python
df = pd.read_excel("Relatório Live Jogo Sassoeiros 11-09-2022.xlsx", skiprows=3, usecols=range(2,13))
df = df.loc[:find_first_nan_row(df)-1]
df
```

```python
# Univariate analysis

col_names = ["Passe Errado", "Desarme", "Interceção de Passe", "Interceção de Remate", "Remate Efetuado", "Golos", "Assistência", "Perda de Bola"]
fig, axes = plt.subplots(2, 4, figsize=(19, 9))

for ax, name in zip(axes.flat, col_names):
    data = df.sort_values(name, ascending=False)
    ax.bar(data["Nome completo"], data[name])
    ax.set_title(name)
    ax.tick_params(axis="x", rotation=45)

plt.suptitle(f"Jogo {df['Data'].iloc[0]} - Atlético vs {df['Adversário'].iloc[0]}")
plt.tight_layout()
plt.show()
```
